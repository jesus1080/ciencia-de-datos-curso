```python
from sklearn import linear_model
import pandas as pd
import numpy as np 
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
```

```python
import sklearn.datasets
dsBoston = sklearn.datasets.load_boston()
dsBoston
```

```python
dsBoston.keys()
```

```python
dfBoston = pd.DataFrame(dsBoston.data, columns=dsBoston.feature_names)
dfBoston.head()
```

```python
dfBoston['MEDV'] = dsBoston.target
dfBoston.head()
```

```python
sns.displot(dfBoston['MEDV'], bins=50)
plt.show()
```

```python
#matriz de correlacion
correlation_matrix_boston = dfBoston.corr().round(2)
correlation_matrix_boston
```

```python
sns.heatmap(correlation_matrix_boston, annot=True)
plt.rcParams['figure.figsize'] = (20,20)
plt.show()
```

```python
#Diagramas de dispersion 
sns.pairplot(dfBoston, height=2)
plt.tight_layout()
```

```python
dfModelBoston = dfBoston [['RM', 'LSTAT', 'PTRATIO']]
dfModelBoston.head()
```

```python
dfModelBostonPrices = dfBoston['MEDV']
dfModelBostonPrices.head()
```

```python
#para crear el modelo devidimos el df en los datos de test y entrenamiento
X_train, X_test, y_train, y_test = train_test_split(dfModelBoston, dfModelBostonPrices, test_size = 0.2, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
lin_model_boston = LinearRegression()
lin_model_boston.fit(X_train, y_train)

print('Coefientes: \n', lin_model_boston.coef_)
print('Termino Independiente: \n', lin_model_boston.intercept_)
```

```python
from sklearn.metrics import r2_score
y_train_predict = lin_model_boston.predict(X_train)
r2 = r2_score(y_train, y_train_predict)
print('The model preformance for training set')
print('--------------------------------------')
print('R2 score is {}'.format(r2))
```

```python
# R2ajustado = 1 - (1-r2)/(n - (p + 1)) VT/(n - 1)
y_test_predict = lin_model_boston.predict(X_test)
r2 = r2_score(y_test, y_test_predict)
print('The model preformance for test set')
print('--------------------------------------')
print('R2 score is {}'.format(r2))

adjusted_r2 = 1 - (1 - r2)*(len(y_test)-1)/(len(y_test)-X_test.shape[1]-1)
print('ajustado R2 score is {}'.format(adjusted_r2))
```

```python
#prediccion del modelo
predicted_data_boston = lin_model_boston.predict(dfModelBoston)
predicted_data_boston[0:5]
```

```python
dfAp = pd.DataFrame({'Actual': dfBoston['MEDV'], 'Prediccion':predicted_data_boston })
dfAp.head(10)
```
