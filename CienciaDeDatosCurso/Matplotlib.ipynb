```python
import matplotlib.pyplot as plt
```

```python
edad_x = [20,25,30,35,40,45,50,55,60,65]
salario_cd_y = [20000,30000,33000,40000,42000,45000,47000,60000,70000,80000]
salario_fed_y = [10000,15000,17000,27000,30000,32000,35000,40000,49000,50000]
salario_bed_y = [15000,20000,25000,27000,35000,40000,45000,50000,55000,60000]
```

```python
plt.plot(edad_x,salario_cd_y)
plt.show()
```

```python
plt.plot(edad_x,salario_cd_y)
plt.title('Salario Medio de un Cientifico de datos por edad')
plt.xlabel('Edad')
plt.ylabel('Salario Medio $USD')
plt.show()
```

```python
plt.plot(edad_x,salario_cd_y,label = 'Cientifico de Datos')
plt.plot(edad_x,salario_fed_y, label = 'Front End Desarrollador')
plt.title('Salario Medio por edad')
plt.xlabel('Edad')
plt.ylabel('Salario Medio $USD')
plt.legend()
plt.show()
```

```python
plt.plot(edad_x,salario_cd_y,label = 'Cientifico de Datos')
plt.plot(edad_x,salario_fed_y, label = 'Front End Desarrollador')
plt.plot(edad_x,salario_bed_y, label = 'Back End Desarrollador')
plt.title('Salario Medio por edad')
plt.xlabel('Edad')
plt.ylabel('Salario Medio $USD')
plt.legend()
plt.show()
```

```python
plt.plot(edad_x,salario_cd_y,label = 'Cientifico de Datos', linewidth=3, marker = '.', markersize=20, linestyle='--', color='red')
plt.plot(edad_x,salario_fed_y, label = 'Front End Desarrollador')
plt.plot(edad_x,salario_bed_y, label = 'Back End Desarrollador')
plt.title('Salario Medio por edad')
plt.xlabel('Edad')
plt.ylabel('Salario Medio $USD')
plt.grid(True)
plt.legend()
plt.show()
```

```python
import numpy as np 
plt.plot(edad_x,salario_cd_y,label = 'Cientifico de Datos')
plt.plot(edad_x,salario_fed_y, label = 'Front End Desarrollador')
plt.plot(edad_x,salario_bed_y, label = 'Back End Desarrollador')
salario_medio = np.sum([salario_cd_y,salario_fed_y,salario_bed_y], axis=0)/3
plt.plot(edad_x, salario_medio, label = 'Salario Medio', linestyle='--')
plt.title('Salario Medio por edad')
plt.xlabel('Edad')
plt.ylabel('Salario Medio $USD')
plt.grid(True)
plt.legend()
plt.show()
```

```python
plt.plot(edad_x,salario_bed_y, label = 'Back End Desarrollador')
salario_medio = np.sum([salario_cd_y,salario_fed_y,salario_bed_y], axis=0)/3
plt.plot(edad_x, salario_medio, label = 'Salario Medio', linestyle='--')

plt.fill_between(edad_x,salario_bed_y,salario_medio,interpolate=True,
                alpha=0.33, label='Between')

plt.title('Salario Medio por edad')
plt.xlabel('Edad')
plt.ylabel('Salario Medio $USD')
plt.grid(True)
plt.legend()
plt.show()
```

```python
plt.plot(edad_x,salario_bed_y, label = 'Back End Desarrollador')
salario_medio = np.sum([salario_cd_y,salario_fed_y,salario_bed_y], axis=0)/3
plt.plot(edad_x, salario_medio, label = 'Salario Medio', linestyle='--')

plt.fill_between(edad_x,salario_bed_y,salario_medio, where=(salario_bed_y>salario_medio),
                interpolate=True,
                alpha=0.33, label='Between')

plt.fill_between(edad_x,salario_bed_y,salario_medio, where=(salario_bed_y<=salario_medio),
                interpolate=True, color='yellow',
                alpha=0.33, label='Between')

plt.title('Salario Medio por edad')
plt.xlabel('Edad')
plt.ylabel('Salario Medio $USD')
plt.grid(True)
plt.legend()
plt.show()
```

```python
plt.plot(edad_x,salario_cd_y,label = 'Cientifico de Datos')
plt.plot(edad_x,salario_fed_y, label = 'Front End Desarrollador')
plt.plot(edad_x,salario_bed_y, label = 'Back End Desarrollador')
salario_medio = np.sum([salario_cd_y,salario_fed_y,salario_bed_y], axis=0)/3
plt.bar(edad_x, salario_medio, label = 'Salario Medio', color='#000')
plt.title('Salario Medio por edad')
plt.xlabel('Edad')
plt.ylabel('Salario Medio $USD')
plt.grid(True)
plt.legend()
plt.show()
```
